```python
from google.cloud import bigquery
import os

# Ruta a tu JSON de credenciales
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "swift-icon-479219-e9-4e210848f274.json"

# Inicializar cliente BigQuery
client = bigquery.Client()

# Dataset que quieres inspeccionar
dataset_id = "swift-icon-479219-e9.Transporte_publico"

# Listar tablas
tables = client.list_tables(dataset_id)
print("Tablas en el dataset:")
for table in tables:
    print(table.table_id)
```

```python
# Ruta al Parquet local
parquet_file = "matches.parquet"

# Tabla destino en BigQuery
table_id = "swift-icon-479219-e9.Transporte_publico.vehiculos"  # cambia 'mi_tabla'

# Configuraci√≥n de carga
job_config = bigquery.LoadJobConfig(
    source_format=bigquery.SourceFormat.PARQUET,
    write_disposition="WRITE_APPEND"  # o "WRITE_TRUNCATE" si quieres reemplazar
)

# Ejecutar carga
with open(parquet_file, "rb") as source_file:
    job = client.load_table_from_file(source_file, table_id, job_config=job_config)

job.result()  # Espera a que termine

print(f"Parquet cargado en {table_id}")
```
