```python
import pandas as pd
```

```python
homologacion_raw_path = "C:\\Users\\hsakurada\\Desktop\\ESTUDIO\\wea\\prueba-tecnica\\CMS-PRUEBA-TECNICA\\utils\\Web_Nomina_Vehiculos_Livianos_y_Medianos_Homologados_29_oct_2025 (1).xls"
```

```python

df = pd.read_excel(homologacion_raw_path, header=[0,1])

nuevas_cols = []

for col0, col1 in df.columns:
    col0 = str(col0).strip()
    col1 = str(col1).strip()

    if "Unnamed" in col1:
        nuevas_cols.append(col0)   # usar solo el nombre bueno
    else:
        nuevas_cols.append(f"{col0} {col1}")

df.columns = nuevas_cols
```
