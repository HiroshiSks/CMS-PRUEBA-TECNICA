```python
import pandas as pd
import re
import unicodedata


# ===========================================================
# 1) NORMALIZADOR ULTRA-RÁPIDO DE TEXTO
# ===========================================================
_re_non_alnum = re.compile(r"[^0-9a-zA-Z]+")
_re_multi_unders = re.compile(r"_+")

def normalize_fast(text: str) -> str:
    """Normaliza texto MUY rápido: sin acentos, minúsculas, underscores limpios."""
    if not isinstance(text, str):
        return text

    # eliminar acentos
    text = unicodedata.normalize("NFKD", text)
    text = "".join(c for c in text if not unicodedata.combining(c))

    # minúsculas
    text = text.lower()

    # reemplazar todo lo raro por _
    text = _re_non_alnum.sub("_", text)
    text = _re_multi_unders.sub("_", text)

    return text.strip("_")
    


# ===========================================================
# 2) EXTRAER CATEGORÍA DE AÑOS DESDE EL NOMBRE DE LA HOJA
#    Formato soportado: "2020 - 2021" | "2020-2021"
# ===========================================================
pat_anhos = re.compile(r"(\d{4})\s*[-–]\s*(\d{4})")

def extraer_categoria_años(nombre_hoja: str) -> str:
    match = pat_anhos.search(nombre_hoja)
    if match:
        a1, a2 = match.groups()
        return f"{a1}_{a2}"
    return None



# ===========================================================
# 3) PROCESAR TODAS LAS HOJAS Y NORMALIZAR COLUMNAS
# ===========================================================
def procesar_homologacion(dfs_normalizados: dict):
    """
    Procesa un diccionario de dataframes ya cargados desde Excel.
    Normaliza columnas, extrae categoría de años y filtra columnas de interés.
    """
    
    columnas_interes = [
        "unnamed_0_level_0",
        "marca",
        "modelo",
        "transmision",
        "combustible",
        "propulsion",
        "cilindrada",
        "emisiones_de_co2_g_km","co2_combinado_g_km",
        "co2_(gr_km)"

    ]

    dfs = {}

    for hoja, df in dfs_normalizados.items():

        # --------------------------------------------------
        # 1) Normalizar columnas usando normalize_fast
        # --------------------------------------------------
        columnas_limpias = []
        for col in df.columns:
            columnas_limpias.append(normalize_fast(col))

        df.columns = columnas_limpias

        # --------------------------------------------------
        # 2) Extraer categoría de años desde el nombre de hoja
        # --------------------------------------------------
        categoria = extraer_categoria_años(hoja)
        df["categoria_años"] = categoria

        # --------------------------------------------------
        # 3) Filtrar columnas
        # --------------------------------------------------
        cols_keep = [c for c in columnas_interes if c in df.columns]
        cols_keep.append("categoria_años")

        df = df[cols_keep]

        dfs[hoja] = df

    return dfs

```
